```python
# Definimos una función para categorizar las horas en morning, afternoon, evening y night

def categorizar_hora(hora):
    if hora >= pd.to_datetime("06:00").time() and hora < pd.to_datetime("12:00").time():
        return 'Morning'
    elif hora >= pd.to_datetime("12:00").time() and hora < pd.to_datetime("18:00").time():
        return 'Afternoon'
    elif hora >= pd.to_datetime("18:00").time() and hora <= pd.to_datetime("23:59").time():
        return 'Evening'
    else:
        return 'Night'
```

```python
# aplicamos la función a la columna Time y creo una nueva Columna Hora_Categorizada
df_filtrado_año['Hora_Categorizada'] = df_filtrado_año["Time"].apply(categorizar_hora)
df_filtrado_año[["Hora_Categorizada"]]
```

```python
replacement_dict = {
    "white": "white shark",
    "grey": "grey shark",
    "blue": "blue shark",
    "tiger": "tiger shark",
    "lemon" : "lemon shark",
    "bull": "bull shark",
    "6'": "6' shark",
    "5'": "5' shark",
    "4'": "4' shark",
    "3'": "3' shark",
    "10'" : "10' shark",
    "12'": "12' shark",
    "8'": "8' shark",
    "7'": "7' shark",
    "2 m": "7' shark",
    "3 m": "10' shark",
    "15 m": "50' shark"
    }

    # Función para reemplazar los tiburones que se parecen
def replace_shark_name(specie):
    if isinstance(specie, str):  # Verifica si es una cadena
        for key, value in replacement_dict.items():
            if key in specie:
                return value
        return specie  # me devuelve el nombre original si no hay coincidencia

# Aplicamos la función a la columna "Species"
df_filtrado_año['Species'] = df_filtrado_año['Species'].apply(replace_shark_name)
```

```python
# quiero crear una columna con las estaciones, según el pais y el mes del año
# como casi todo son paises y los oceanos solo tienen 7 y 11 filas no se verá muy afectado
```

```python
hemisferio_sur = ["Australia", "Brazil", "South Africa", "New Zealand", "New Caledonia", 
                  "Solomon Islands", "Reunion", "French polynesia", "Mozambique", "Fiji", 
                  "Papua new guinea", "Jamaica","Tonga", "Vanuatu", "Marshall islands"]

hemsiferio_norte = ["Usa", "Mexico", "England", "Costa rica", "Bahamas", "Cuba", "Spain", 
                    "Japan", "Egypt", "United kingdom", "Philippines", "Vietnam", "Panama",
                      "Croatia", "Italy", "India", "Hong kong", "Iran", "Greece", "Turkey"]
```

```python
def estacion(fila):
    country = fila["Country"]
    mes = fila["Mes"]

    if country in hemisferio_sur:
        if mes in ["Dicember", "January", "February"]:
            return "Summer"
        elif mes in ["March", "April", "May"]:
            return "Fall"
        elif mes in ["June", "July", "August"]:
            return "Winter"
        else:
            return "Spring"
    else: # hemisferio norte
        if mes in ["Dicember", "January", "February"]:
            return "Winter"
        elif mes in ["March", "April", "May"]:
            return "Spring"
        elif mes in ["June", "July", "August"]:
            return "Summer"
        else:
            return "Fall"
```

```python
df_filtrado_año["Season"]= df_filtrado_año.apply(estacion, axis =1)
```
